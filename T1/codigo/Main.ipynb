```python
import numpy as np

def print_matrix(matrix):
    # Get the maximum width needed for any element
    max_width = max(len(str(element)) for row in matrix for element in row)
    
    for line in matrix:
        # Format each element to have consistent width
        formatted_row = [str(element).rjust(max_width) for element in line]
        print(" ".join(formatted_row))

# Contruindo a Malha
h = 0.25

x = np.arange(0,1+h, h)
y = np.arange(-1,1+h, h)


print(x)
print(y)

n = len(x)
m = len(y)

print(n)
print(m)


```
